```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("GlobalTemperatures.csv", parse_dates=["dt"])
dataset = dataset.loc[:,["dt","LandAverageTemperature"]]
dataset = dataset.dropna(axis=0)
dataset.columns = ["Date","LandAverageTemperature"] 
print(dataset)
```

```python
import datetime as dt
dataset['Date'] = pd.to_datetime(dataset['Date'])
dataset['Date']=dataset['Date'].map(dt.datetime.toordinal)
```

```python
print(dataset.iloc[-1:-20:-1])
```

```python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
X = sc_X.fit_transform(dataset["Date"].values.reshape(-1,1))

sc_y = StandardScaler()
y = sc_y.fit_transform(dataset["LandAverageTemperature"].values.reshape(-1,1))
y = y.reshape(-1)
print(y)
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf', gamma='auto')
regressor.fit(X_train,y_train)
```

```python
y_pred = sc_y.inverse_transform( regressor.predict(sc_X.transform(np.array( X_test.reshape(-1,1) ))))
```

```python
print(y_pred)
```

```python
print(y_test)
```

```python
import sklearn.metrics as sm
print("R2 score =", round(sm.r2_score(y_test, y_pred), 2))
```
